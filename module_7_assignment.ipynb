```python
# ans 3
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.show()
```

```python
# ans 4

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

np.random.seed(0)
data = np.random.multivariate_normal([0, 0], [[1, 0.7], [0.7, 1]], size=100)
df = pd.DataFrame(data, columns=['X', 'Y'])

corr_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.show()
```

```python
# ans 5
import plotly.graph_objects as go

categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

fig = go.Figure(data=[go.Bar(x=categories, y=values)])
fig.update_layout(title='Bar Plot', xaxis_title='Category', yaxis_title='Value')
fig.show()
```

```python
# ans 6
import pandas as pd

data = {'Name': ['John', 'Anna', 'Peter', 'Linda'],
        'Age': [28, 24, 35, 32]}
df = pd.DataFrame(data)
df['Adult'] = df['Age'].apply(lambda x: 'Yes' if x >= 18 else 'No')
print(df)
```

```python
# ans 8
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Plot')
plt.legend()
plt.show()
```

```python
# ans 10
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(0, 1, 1000)

plt.figure(figsize=(8, 6))
sns.histplot(data, bins=20, kde=True)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
